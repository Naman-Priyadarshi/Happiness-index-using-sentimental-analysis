```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Tweets_Sentiment_Analysis_RoBERTa_Raw_Values.csv'  
df = pd.read_csv(file_path)

# Select only the 'negative', 'positive', and 'neutral' columns
selected_columns = ['negative', 'positive', 'neutral']
df_selected = df[selected_columns]

# Find the column with the maximum value for each row
df['max_sentiment'] = df_selected.idxmax(axis=1)

# Count the occurrences of each max sentiment
sentiment_counts = df['max_sentiment'].value_counts()

# Plot the bar graph
fig, ax = plt.subplots()
sentiment_counts.plot(kind='bar', ax=ax, color=['red', 'gray', 'green'])

# Set labels and title
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Maximum Sentiment Analysis')

# Show the plot
plt.show()
```
