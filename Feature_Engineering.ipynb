# Feature Engineering

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import textstat

# Load your dataset
df = pd.read_csv("uttarakhand_clean.csv")

# Handle NaN values in 'preprocessed_text' column
df['preprocessed_text'] = df['text'].fillna('')

# Text Length
df['text_length'] = df['text'].apply(lambda x: len(str(x)))

# Word Count
df['word_count'] = df['text'].apply(lambda x: len(str(x).split()))

# Emotional Intensity
df['emotional_intensity'] = df['text'].apply(lambda x: str(x).count('!'))

# N-grams
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))
ngrams = ngram_vectorizer.fit_transform(df['text'].astype(str))

# Readability Metrics
df['readability'] = df['text'].apply(lambda x: textstat.flesch_kincaid_grade(str(x)))

# Display the updated DataFrame
print(df)

df.to_csv("uttarakhand_with_features.csv", index=False)
```
